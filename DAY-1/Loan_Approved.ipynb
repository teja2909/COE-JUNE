#### Handling Missing Values

```python
for col in ['Gender', 'Married', 'Dependents', 'Self_Employed','Loan_Amount_Term', 'Credit_History']:
    df[col].fillna(df[col].mode()[0], inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Create a box plot for each numerical column
plt.figure(figsize=(14, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, (len(numeric_cols) + 1) // 2, i)
    sns.boxplot(data=df, y=col)
    plt.title(f"Box Plot of {col}")
    plt.tight_layout()

plt.show()
```

```python
cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

outliers_dict = {}

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower) | (df[col] > upper)].index.tolist()
    
    outliers_dict[col] = outliers

for col, outlier_ids in outliers_dict.items():
    print(f"{col}: {len(outlier_ids)} outliers")
```

```python
cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

# Create subplots for boxplots
plt.figure(figsize=(12, 5 * len(cols)))

for i, col in enumerate(cols, 1):
    plt.subplot(len(cols), 1, i)
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot of {col}")

plt.tight_layout()
plt.show()
```

```python
print(df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']].isnull().sum())

df['ApplicantIncome'].fillna(df['ApplicantIncome'].median(), inplace=True)
df['CoapplicantIncome'].fillna(df['CoapplicantIncome'].median(), inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)

print("âœ… Missing values handled.")
print(df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']].isnull().sum())
```
