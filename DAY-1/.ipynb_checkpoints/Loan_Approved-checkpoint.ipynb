#### Handling Missing Values

```python
for col in ['Gender', 'Married', 'Dependents', 'Self_Employed','Loan_Amount_Term', 'Credit_History']:
    df[col].fillna(df[col].mode()[0], inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

## Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Applicant Income vs Loan Amount

```python
sns.scatterplot(data=df, x="ApplicantIncome", y="LoanAmount", hue="Loan_Status (Approved)")
plt.title("Applicant Income vs Loan Amount")
plt.show()
```

#### Coapplicant Income vs Loan Amount

```python
sns.scatterplot(data=df, x="CoapplicantIncome", y="LoanAmount", hue="Loan_Status (Approved)")
plt.title("Coapplicant Income vs Loan Amount")
plt.show()
```

#### Loan Amount by Education Level

```python
sns.boxplot(data=df, x="Education", y="LoanAmount")
plt.title("Loan Amount by Education Level")
plt.show()
```

#### Loan Approval by Property Area

```python
sns.countplot(data=df, x="Property_Area", hue="Loan_Status (Approved)")
plt.title("Loan Approval by Property Area")
plt.show()
```

#### Loan Approval by Gender

```python
sns.countplot(data=df, x="Gender", hue="Loan_Status (Approved)")
plt.title("Loan Approval by Gender")
plt.show()
```

#### Loan Approval by Credit History

```python
sns.countplot(data=df, x="Credit_History", hue="Loan_Status (Approved)")
plt.title("Loan Approval by Credit History")
plt.show()
```

#### Loan Amount Distribution

```python
sns.histplot(data=df, x="LoanAmount", bins=30, kde=True)
plt.title("Loan Amount Distribution")
plt.show()
```

#### Loan Approval by Number of Dependents

```python
sns.countplot(data=df, x="Dependents", hue="Loan_Status (Approved)")
plt.title("Loan Approval by Number of Dependents")
plt.show()
```
